# 📌 Apuntes de Git & GitHub

---

# 1️⃣ ¿Qué es Git?

**Git** es un **sistema de control de versiones (local)**.

Sirve para:

- 💾 Guardar cambios en tu código  
- ⏪ Volver a versiones anteriores  
- 👥 Trabajar en equipo sin perder información  
- 🧾 Saber quién hizo cada cambio  

---

# 2️⃣ ¿Qué es GitHub?

**GitHub** es una **plataforma online (remota)** donde puedes:

- ☁️ Subir tus repositorios Git  
- 👥 Compartir código  
- 🤝 Colaborar en proyectos  
- 📦 Tener copia de seguridad en la nube  

## Git vs GitHub

| Git | GitHub |
|----|----|
| Local | En la nube |
| Control de versiones | Plataforma para alojar repos |
| Funciona sin internet | Necesita internet |

⚠️ **GitHub usa Git, pero no son lo mismo.**

---

# 3️⃣ Conceptos básicos

## 📁 Repositorio

Un **repositorio** es una **carpeta de proyecto controlada por Git**.

Se crea con:

git init


Esto convierte una carpeta normal en un **repositorio Git**.

---

## 📸 Commit

Un **commit** es una **foto del proyecto en un momento concreto**.

Cada commit tiene:

- Identificador único  
- Autor  
- Fecha  
- Mensaje descriptivo  

Ejemplo:

git commit -m "Añade página de login"


---

## 🟢 Stage (Área de preparación)

Antes de hacer un commit debemos **añadir los archivos al área de preparación (stage)**.

Añadir un archivo:

git add archivo.py


Añadir todos los archivos:

git add .


---

## 🔄 Flujo de Git

Working Directory
↓
git add
↓
Stage
↓
git commit
↓
Historial Git


---

## 🌿 Branch (Rama)

Una **branch (rama)** es una **versión paralela del proyecto**.

Permite:

- Crear nuevas funcionalidades  
- Probar cambios  
- No afectar la rama principal  

La rama principal normalmente se llama:

main


---

# 4️⃣ Crear y gestionar ramas

## Ver en qué rama estás

git branch


Ejemplo:

main


El `*` indica la **rama actual**.

---

## Crear una rama

git branch nueva-rama


⚠️ Solo crea la rama, **no cambia a ella**.

---

## Cambiar de rama

git checkout nueva-rama


---

## Crear y cambiar de rama en un solo paso

git checkout -b login


---

## Forma moderna recomendada por Git

git switch -c dashboard


---

# 5️⃣ Merge (unir ramas)

El **merge** sirve para **unir los cambios de una rama con otra**.

git merge nombre-rama


Esto mezcla el contenido de la rama con la rama actual.

---

# 6️⃣ Flujo básico de trabajo

## Crear un proyecto desde cero

git init
git add .
git commit -m "Primer commit"


---

## Subir proyecto a GitHub

### 1️⃣ Crear repositorio en GitHub

Primero creamos un repositorio en la web de GitHub.

---

### 2️⃣ Conectar con el repositorio local

git remote add origin URL_DEL_REPO


---

### 3️⃣ Subir el proyecto

git push -u origin main


El `-u` vincula la rama local con la remota.

---

# 7️⃣ Clonar un repositorio

Clonar significa **descargar un repositorio existente**.

git clone URL_DEL_REPO


Esto crea una carpeta con el proyecto.

---

# 8️⃣ Comandos más importantes

| Comando | Función |
|------|------|
| git init | Iniciar repositorio |
| git status | Ver estado |
| git add . | Añadir cambios |
| git commit -m | Guardar cambios |
| git log | Ver historial |
| git branch | Ver ramas |
| git checkout | Cambiar rama |
| git merge | Unir ramas |
| git push | Subir cambios |
| git pull | Descargar cambios |

---

# 9️⃣ Diferencia entre Push y Pull

## 🔼 Push

Sube los cambios **del ordenador a GitHub**.

git push


---

## 🔽 Pull

Descarga los cambios **de GitHub al ordenador**.

git pull


---

# 🔟 Conflictos

Los **conflictos** ocurren cuando:

- Dos personas modifican **la misma línea**
- Se intenta hacer un **merge con cambios incompatibles**

Git pedirá elegir qué versión conservar.

---

# 1️⃣1️⃣ Buenas prácticas

✔ Hacer commits pequeños  
✔ Escribir mensajes claros  
✔ Crear ramas para nuevas funcionalidades  
✔ No subir archivos innecesarios (`.gitignore`)  
✔ Hacer `git pull` antes de `git push`
