# 1️⃣ ¿Qué es HTML?

## 1.1 Definición

**HTML** significa **HyperText Markup Language**  
(Lenguaje de Marcado de Hipertexto).

Es el **lenguaje estándar utilizado para crear páginas web**.

---

## 1.2 Características principales

- Define la **estructura de una página web**
- Organiza el contenido (**texto, imágenes, enlaces, tablas…**)
- ❌ No es un lenguaje de programación
- Funciona mediante **etiquetas**

---

# 2️⃣ ¿Cómo funciona HTML?

## 2.1 Las etiquetas (tags)

HTML utiliza **etiquetas** para estructurar el contenido.

Ejemplo:

<p>Este es un párrafo</p> ```
La mayoría de etiquetas tienen:

Etiqueta de apertura → <p>

Contenido

Etiqueta de cierre → </p>

2.2 Etiquetas sin cierre
Algunas etiquetas no necesitan cierre:

<img>
<br>
<hr>
3️⃣ Estructura básica de un documento HTML
Todo documento HTML tiene una estructura base:

<!DOCTYPE html>
<html>

<head>
    <title>Título de la página</title>
</head>

<body>
    <h1>Mi primera página</h1>
    <p>Contenido aquí</p>
</body>

</html>
3.1 Partes principales
Etiqueta	Función
<!DOCTYPE html>	Indica que es un documento HTML5
<html>	Contiene todo el documento
<head>	Información interna (no visible)
<body>	Contenido visible de la página   
   
4️⃣ Etiquetas más importantes
4.1 Encabezados
Los encabezados van desde h1 hasta h6.

<h1>Encabezado principal</h1>
<h2>Subtítulo</h2>
<h3>Subapartado</h3>
<h1> → título principal

<h2> → subtítulo

<h3> → subsección

4.2 Párrafos
Los párrafos se crean con la etiqueta:

<p>Texto de un párrafo</p>
4.3 Enlaces
Los enlaces se crean con la etiqueta <a>.

<a href="https://www.google.com">Ir a Google</a>
Atributos:

href → dirección del enlace

4.4 Imágenes
Para insertar imágenes se usa:

<img src="imagen.jpg" alt="Descripción">
Atributos:

src → ruta de la imagen

alt → texto alternativo (accesibilidad)

4.5 Listas
Lista ordenada
<ol>
    <li>Elemento 1</li>
    <li>Elemento 2</li>
</ol>
Lista desordenada
<ul>
    <li>Elemento A</li>
    <li>Elemento B</li>
</ul>
5️⃣ Atributos en HTML
Los atributos proporcionan información adicional a las etiquetas.

Ejemplo:

<a href="https://www.ejemplo.com" target="_blank">

Atributo	Función
href	Dirección del enlace
target="_blank"	Abre el enlace en una nueva pestaña   

6️⃣ Comentarios en HTML
Los comentarios sirven para añadir notas que no se ven en la página web.   

<!-- Esto es un comentario -->

7️⃣ HTML y CSS
Cada tecnología tiene una función diferente.

Tecnología	Función
HTML	Estructura
CSS	Diseño
JavaScript	Interactividad
Ejemplo de conexión con CSS
<link rel="stylesheet" href="estilos.css">
Esto conecta el archivo HTML con una hoja de estilos CSS.   

8️⃣ Buenas prácticas en HTML
✔ Usar etiquetas semánticas correctamente
✔ Cerrar siempre las etiquetas
✔ Escribir código ordenado y con sangría
✔ Usar nombres claros en archivos e imágenes

🧠 Resumen conceptual
HTML = Estructura

CSS = Diseño

JavaScript = Interactividad
HTML crea la base de la página web,
CSS la hace visualmente atractiva,
y JavaScript añade comportamiento e interacción.
